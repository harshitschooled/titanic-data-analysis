```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

numerical_col=['Age','Experience_Years','Monthly_Salary']

# Drop rows with NaN values in numerical columns and reset index for clean assignment
df.dropna(subset=numerical_col,inplace=True)
df.reset_index(drop=True, inplace=True)

minmax_scaler=MinMaxScaler()
standard_scaler=StandardScaler()

# Apply Min-Max Scaling and Standard Scaling to 'Age'
df['Age_MinMax']=minmax_scaler.fit_transform(df[['Age']])
df['Age_Standard']=standard_scaler.fit_transform(df[['Age']])

# Apply Min-Max Scaling and Standard Scaling to 'Experience_Years'
df['Experience_Years_MinMax']=minmax_scaler.fit_transform(df[['Experience_Years']])
df['Experience_Years_Standard']=standard_scaler.fit_transform(df[['Experience_Years']])

# Apply Min-Max Scaling and Standard Scaling to 'Monthly_Salary'
df['Monthly_Salary_MinMax']=minmax_scaler.fit_transform(df[['Monthly_Salary']])
df['Monthly_Salary_Standard']=standard_scaler.fit_transform(df[['Monthly_Salary']])

# Print columns to verify creation of scaled columns
print("DataFrame columns after scaling:")
print(df.columns)
```
